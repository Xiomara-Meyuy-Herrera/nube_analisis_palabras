```python
import pandas as pd
import numpy as np
```

```python
import numpy as np

def generate_cardiovascular_risk_dataset(num_patients=100):
    """
    Generates a synthetic dataset of possible patients with cardiovascular risk factors.

    Args:
        num_patients (int): The number of patients to generate.

    Returns:
        pd.DataFrame: A DataFrame containing synthetic patient data and risk levels.
    """
    np.random.seed(42)  # for reproducibility

    # Generate synthetic data for relevant features
    age = np.random.randint(30, 80, num_patients)
    sex = np.random.choice(['Male', 'Female'], num_patients)
    cholesterol = np.random.uniform(150, 300, num_patients)
    blood_pressure_systolic = np.random.uniform(100, 180, num_patients)
    blood_pressure_diastolic = np.random.uniform(60, 120, num_patients)
    bmi = np.random.uniform(18, 35, num_patients)
    smoking = np.random.choice([0, 1], num_patients, p=[0.7, 0.3]) # 30% smokers
    diabetes = np.random.choice([0, 1], num_patients, p=[0.85, 0.15]) # 15% diabetic

    # Determine cardiovascular risk (simplified logic for demonstration)
    # A simple risk score based on common risk factors
    risk_score = (age * 0.1) + (cholesterol * 0.05) + (blood_pressure_systolic * 0.05) + \
                 (blood_pressure_diastolic * 0.05) + (bmi * 0.1) + (smoking * 2) + (diabetes * 3)

    # Assign risk levels based on risk score
    risk_level = pd.cut(risk_score, bins=[0, 20, 35, np.inf], labels=['Low', 'Medium', 'High'])

    # Create a DataFrame
    data = {
        'Age': age,
        'Sex': sex,
        'Cholesterol': cholesterol,
        'Blood_Pressure_Systolic': blood_pressure_systolic,
        'Blood_Pressure_Diastolic': blood_pressure_diastolic,
        'BMI': bmi,
        'Smoking': smoking,
        'Diabetes': diabetes,
        'Risk_Score': risk_score,
        'Risk_Level': risk_level
    }
    df = pd.DataFrame(data)

    return df

# Example usage:
df_patients = generate_cardiovascular_risk_dataset(num_patients=200)
display(df_patients.head())
```
